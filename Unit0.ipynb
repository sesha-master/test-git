```
def get_median(list1):
    list1.sort()
    l = len(list1)
    if (l%2==0): return (list1[l//2-1]+list1[l//2])/2
    else: return list1[l//2]
print(get_median([10, 16, 22, 23, 24 ]))

print(15 // 4)
```

```
import re 


def count_letters(sentence, average=False):
    res = 0
    if average:
        pattern = re.compile('\S+')
        res1 = pattern.findall(sentence)
        for i in res1:
          res += len(i)
        res = res / len(res1)
    else:
        pattern = re.compile('\s')
        res = len(sentence) - len(pattern.findall(sentence))
    return res
count_letters("I will build my own theme park", average=True) # => 24

    
```
